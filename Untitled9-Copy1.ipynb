```python
#############################################################################################BACKTRACK#####################################################################################################
##########################################################################Les résultats seront envoyés sur un fichier excel################################################################################

#Fonction qui renverra un tableau de toutes les dataframe de chaque monnaie et leur indicateur

def ichimokucloud(currencies, interval, s_date, e_date): 
    currencies_data=[]*len(currencies)
    for i in range(len(currencies)):
        frame = pd.DataFrame(client.get_historical_klines(currencies[i], interval, s_date, e_date)) #variable temporaire contenant le dataframe de la monnaie
        frame = frame.iloc[:,:6]  #Filtre des infos
        print(currencies[i])
        frame.columns = ['Time','Open', 'High', 'Low', 'Close', 'Volume']  #Nom des colonnes
        #frame = frame.set_index('Time')    #L'index devient la colonne temps
        frame = frame.astype(float)      #Transformer toutes les données en float
        frame.Time = pd.to_datetime(frame.Time, unit='ms')    #affichage du temps en notre système horaire
        
        
        
        
        ichimoku = frame.copy()
    
        high_prices = ichimoku['High']
        close_prices = ichimoku['Close']
        low_prices = ichimoku['Low']
        #dates = ichimoku['Time']  # contains datetime objects

        # Tenkan-sen (Conversion Line): (9-period high + 9-period low)/2))
        period9_high = high_prices.rolling(window=9).max()
        period9_low = low_prices.rolling(window=9).min()
        ichimoku['Tenkan_Sen'] = (period9_high + period9_low) / 2

        # Kijun-sen (Base Line): (26-period high + 26-period low)/2))
        period26_high = high_prices.rolling(window=26).max()
        period26_low = low_prices.rolling(window=26).min()
        ichimoku['Kijun_Sen'] = (period26_high + period26_low) / 2

        # Senkou Span A (Leading Span A): (Conversion Line + Base Line)/2))
        ichimoku['SPA'] = ((ichimoku['Tenkan_Sen'] + ichimoku['Kijun_Sen']) / 2).shift(26)

        # Senkou Span B (Leading Span B): (52-period high + 52-period low)/2))
        period52_high = high_prices.rolling(window=52).max()
        period52_low = low_prices.rolling(window=52).min()
        ichimoku['SPB'] = ((period52_high + period52_low) / 2).shift(26)

        #Récupère la date du listing de la monnaie et vérifie si elle est antérieure a la date de début de simulation. Si ce n'est pas le cas il faut créé un Dataframe a ajouter pour qu'ils aient tous la même longueur
        c_date=ichimoku.Time[0]
        sim_date=s_date
        sim_date=datetime.datetime.strptime(sim_date, '%d %b, %Y')
        sim_date=datetime.datetime.strftime(sim_date,'%Y-%m-%d')
        sim_hour=' 00:00:00'
        sim_date=sim_date+sim_hour
        sim_date= datetime.datetime.strptime(sim_date, '%Y-%m-%d %H:%M:%S')
        diff=(c_date-sim_date).total_seconds()
        diff=diff/14400
        #Transforme cette différence en nombre de période de 4h
        diff=int(diff)
        #Crée un autre df avec des NaN avec le nombre de période de différence 
        dummyarray = np.empty((diff,10))
        dummyarray[:] = np.nan
        o = pd.DataFrame(dummyarray)
        o.columns = ['Time','Open', 'High', 'Low', 'Close', 'Volume','Tenkan_Sen','Kijun_Sen','SPA','SPB']
        #Concatenate les deux Dataframes
        ichimoku=pd.concat([o,ichimoku],ignore_index=True)
        #Condition a rajouter car au final les Dataframe n'avaient pas tous la même longueur, je penses due à un problème lors du calcul du nombre de périodes de différence
        if (len(ichimoku!=2191)):
            diff2=2191-len(ichimoku)
            dummyarray = np.empty((diff2,10))
            dummyarray[:] = np.nan
            o = pd.DataFrame(dummyarray)
            o.columns = ['Time','Open', 'High', 'Low', 'Close', 'Volume','Tenkan_Sen','Kijun_Sen','SPA','SPB']
            ichimoku=pd.concat([o,ichimoku],ignore_index=True)
        currencies_data.append(ichimoku)     #Ajouter une copie de la variable temporaire dans le tableau currencies_data   
    currencies_data[1].to_excel(r'C:\Users\User\test0001.xlsx', index=True, engine='xlsxwriter')
    return(currencies_data)


#Prend en parametre les dataframes de chaque monnaie avec leur indicateur et renvoie le tableau des trades

def backtrack(currencies,interval,start_date,end_date,risk_per_trade,capital):
    Symbol=[]
    Size=[]
    Open=[]
    Open_Time=[]
    Close=[]
    Close_Time=[]
    Current_value=[]
    Profit_pct=[]
    Profit_dollar=[]
    pos=[0]*len(currencies)
    num=[0]*len(currencies)
    ntrade=0
    risk_per_trade=float(risk_per_trade)
    capital=float(capital)
    
    
    
    currencies_data=ichimokucloud(currencies, interval, start_date,end_date)
    l=len(currencies)
    l=l-1
    
    for j in range(len(currencies_data[0])-1):
        for i in range(0,l):
            if(pos[i]==0):
                if(np.logical_and(currencies_data[i].Close[j]>currencies_data[i].SPB[j],currencies_data[i].Tenkan_Sen[j]>currencies_data[i].Kijun_Sen[j])==True): #règle pour l'indicateur 
                    
                    Open_Time.append(currencies_data[i].iloc[j+1][0])
                    Close_Time.append(0)
                    
                    
                    Open.append(currencies_data[i].iloc[j+1].Close)
                    Close.append(0)
                    
                    Symbol.append(currencies[i])
                    Size.append((risk_per_trade/100)*capital)
                    
                    Profit_pct.append(0)
                    Profit_dollar.append(0)
                    
                    capital=capital*(1-(risk_per_trade/100))
                    num[i]=ntrade
                    pos[i]=1
                    ntrade=ntrade+1
                    
            elif(pos[i]==1):
                if(currencies_data[i].Close[j]<currencies_data[i].SPA[j]):
                    Close_Time[num[i]]=currencies_data[i].iloc[j+1][0]
                    Close[num[i]]=currencies_data[i].iloc[j+1].Close
                    
                    Profit_pct[num[i]]=((Close[num[i]]/Open[num[i]])-1)*100
                    Profit_dollar[num[i]]=Size[num[i]]*((currencies_data[i].iloc[j+1].Close)/(Open[num[i]]))-Size[num[i]]
                    
                    capital=capital+Size[num[i]]+Profit_dollar[num[i]]                    
                    pos[i]=0
                    num[i]=0
                elif():
                    Current_value[num[i]]=Size[num[i]]*((currencies_data[i].iloc[j+1].Close)/(Open[num[i]]))
    trades={'Symbol': Symbol, 'Size': Size, 'Open':Open, 'Open_Time':Open_Time, 'Close': Close, 'Close_Time': Close_Time,'Current_value': Current_value, 'Profit_pct': Profit_pct,'Profit_dollar': Profit_dollar }
    trades = pd.DataFrame.from_dict(trades, orient='index')
    trades = trades.transpose()   
    trades.to_excel(r'C:\Users\User\bot3.xlsx', index=True, engine='xlsxwriter')
    return(trades)
```

```python
#Fonction qui transforme les données d'une monnaie au même format que les données des précèdentes périodes
def transform_live_data(msg):
    df=pd.DataFrame([msg])
    df=df.data
    df=df[0]
    frame=df['k']
    frame=pd.DataFrame(frame, index=[0])
    clist=['o','h', 'l', 'c','x']
    frame=frame[clist]
    frame.columns=['Open','High','Low', 'Close','Candle']
    frame=frame.astype(float)
    frame.insert(0,'Time',df['E'])
    frame.insert(0,'Symbol',df['s'])
    frame.Time=pd.to_datetime(frame.Time, unit='ms')  
    frame=pd.DataFrame(frame)
    return(frame)
```
